```python
 def categorizar_idade(TP_FAIXA_ETARIA):
    if 1 <= TP_FAIXA_ETARIA <= 10:
        return '1-25'
    elif TP_FAIXA_ETARIA == 11:
        return '26-30'
    elif 12 <= TP_FAIXA_ETARIA <= 13:
        return '31-40'
    elif 14 <= TP_FAIXA_ETARIA <= 15:
        return '41-50'
    elif 16 <= TP_FAIXA_ETARIA <= 17:
        return '51-60'
    elif 18 <= TP_FAIXA_ETARIA <= 20:
        return '61 a mais'
    else:
        return None

# Criando a nova coluna com as categorias
DADOS['idade_2'] = DADOS['TP_FAIXA_ETARIA'].apply(categorizar_idade)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#sns.boxplot(data=DADOS, x='idade_2', y='NU_NOTA_CN')
# Criando a figura com 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
fig.suptitle('Comparação de Notas por Idade (Idade_2)', fontsize=16)

# Gráfico 1: Notas de Ciências da Natureza
sns.boxplot(data=DADOS, x='idade_2', y='NU_NOTA_MT', ax=axes[0, 0])
axes[0, 0].set_title('Matemática')

# Gráfico 2: Notas de Matemática
sns.boxplot(data=DADOS, x='idade_2', y='NU_NOTA_LC', ax=axes[0, 1])
axes[0, 1].set_title('Linguagem e códigos')

# Gráfico 3: Notas de Linguagens
sns.boxplot(data=DADOS, x='idade_2', y='NU_NOTA_CN', ax=axes[1, 0])
axes[1, 0].set_title('Ciencias da Natureza')

# Gráfico 4: Notas de Humanas
sns.boxplot(data=DADOS, x='idade_2', y='NU_NOTA_CH', ax=axes[1, 1])
axes[1, 1].set_title('Ciencias Humanas')

# Ajustando o layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

#2.Analisar correlações de variáveis que identificar dentro do dataset com a variável dependente nota total (NU_NOTA_CN NU_NOTA_CH, NU_NOTA_LC, NU_NOTA_M.T)

```python
#Aumentando uma coluna com as medias das notas
#Removendo as colunas que não se fará a correlação
DADOS1 = DADOS.drop('NO_MUNICIPIO_ESC', axis=1)

DADOS1 = DADOS1.drop('CO_UF_ESC', axis=1)
DADOS1 = DADOS1.drop('SG_UF_ESC', axis=1)
DADOS1 = DADOS1.drop('NU_INSCRICAO', axis=1)
DADOS1 = DADOS1.drop('NO_MUNICIPIO_PROVA', axis=1)
DADOS1 = DADOS1.drop('CO_MUNICIPIO_PROVA', axis=1)
DADOS1 = DADOS1.drop('CO_UF_PROVA', axis=1)
DADOS1 = DADOS1.drop('SG_UF_PROVA', axis=1)
DADOS1 = DADOS1.drop('CO_PROVA_CN', axis=1)
DADOS1 = DADOS1.drop('CO_PROVA_CH', axis=1)
DADOS1 = DADOS1.drop('CO_PROVA_MT', axis=1)
DADOS1 = DADOS1.drop('CO_PROVA_LC', axis=1)
DADOS1 = DADOS1.drop('TX_RESPOSTAS_CN', axis=1)
DADOS1 = DADOS1.drop('TX_RESPOSTAS_CH', axis=1)
DADOS1 = DADOS1.drop('TX_RESPOSTAS_MT', axis=1)
DADOS1 = DADOS1.drop('TX_RESPOSTAS_LC', axis=1)
DADOS1 = DADOS1.drop('TX_GABARITO_CN', axis=1)
DADOS1 = DADOS1.drop('TX_GABARITO_CH', axis=1)
DADOS1 = DADOS1.drop('TX_GABARITO_MT', axis=1)
DADOS1 = DADOS1.drop('TX_GABARITO_LC', axis=1)
DADOS1 = DADOS1.drop('Q001', axis=1)
DADOS1 = DADOS1.drop('TP_SEXO', axis=1)
# Remover as últimas 25
DADOS1 = DADOS1.drop(DADOS1.columns[-26:], axis=1)
#DADOS1['TP_SEXO'] = DADOS1['TP_SEXO'].map({'Masculino': 1, 'Feminino': 0})
DADOS1['MEDIA_NOTAS'] = DADOS1[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']].mean(axis=1)
DADOS1 = DADOS1.apply(pd.to_numeric, errors='coerce')
```

```python
# Calculando a correlação
correlations = DADOS1.corr()

# Visualizando a matriz de correlação com textos reduzidos
plt.figure(figsize=(12, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.0f', vmin=-1, vmax=1,
            xticklabels=correlations.columns, yticklabels=correlations.columns)

# Ajustando o tamanho da fonte e a rotação das etiquetas
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotaciona as etiquetas no eixo X
plt.yticks(rotation=0, fontsize=10)  # Rotaciona as etiquetas no eixo Y

plt.title('Matriz de Correlação')
plt.show()
```

#5.Gerar visualizações (Data viz) que melhor estratifiquem e demonstremos dados do bloco de DADOS DO QUESTIONÁRIO SOCIOECONÔMICO, verificando o comportamento das notas 4 provas vs. estes dados.

```python
DADOS3 = DADOS.iloc[:, -27:]
# Criando um dicionário para mapear as letras para os números
mapa_letras_para_numeros = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6,
                            'G': 7, 'H': 8, 'I': 9, 'J': 10, 'K': 11, 'L': 12,
                            'M': 13, 'N': 14, 'O': 15, 'P': 16, 'Q': 17}

# Gerando a lista de colunas dinamicamente
colunas = [f"Q{str(i).zfill(3)}" for i in range(1, 26)]  # Q001 a Q025

# Usando um loop para substituir as letras pelos números
for coluna in colunas:
    if coluna in DADOS3.columns:  # Verifica se a coluna existe na tabela
        DADOS3[coluna] = DADOS3[coluna].replace(mapa_letras_para_numeros)
```

```python
DADOS3 = DADOS3.iloc[:, -27:]
DADOS4 =  DADOS3.iloc[:, :25]
DADOS6=DADOS[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC','NU_NOTA_MT']]
DADOS6
DADOS7=pd.concat([DADOS6, DADOS4], axis=1)
DADOS7
```

```python
# Calculando a correlação
correlations = DADOS7.corr()

# Visualizando a matriz de correlação com textos reduzidos
plt.figure(figsize=(12, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.0f', vmin=-1, vmax=1,
            xticklabels=correlations.columns, yticklabels=correlations.columns)

# Ajustando o tamanho da fonte e a rotação das etiquetas
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotaciona as etiquetas no eixo X
plt.yticks(rotation=0, fontsize=10)  # Rotaciona as etiquetas no eixo Y

plt.title('Matriz de Correlação')
plt.show()
```
